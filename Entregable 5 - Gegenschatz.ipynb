```python
# Importamos las librerías tradicionales
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Importamos las librerías relacionadas con json
import json

# Importamos las librerías relacionadas con llamadas HTTPS
import requests
```

```python
url = "https://jsonplaceholder.typicode.com/users"

# Hacemos un llamado https de tipo get
respuesta = requests.get(url)
texto = respuesta.text

# Armamos el DataFrame
jsondata = json.loads(texto)
df = pd.DataFrame.from_dict(jsondata)
df.head()

# Seteamos el índice para que quede mejor
df = df.set_index('id')
df.head()
```

Pero hay un problema, los datos de las columnas address y company son diccionarios en si mismos.
Entonces, usamos ciclos for para generar dos dataframes separados con las direcciones y las compañías.
Luego, los joineamos con un pd.concat

```python
# Arrancamos por las direcciones
address_dict = pd.Series.to_dict(df['address'])
```

```python
users_ids = []
address_df = pd.DataFrame(columns = ['id', 'street', 'suite', 'city', 'zipcode', 'lat', 'long'])
address_df
```

```python
for user_id, d in address_dict.items(): 
    address_df = address_df.append({'id' : user_id,
                            'street' : d['street'], 
                            'suite' : d['suite'],
                            'city' : d['city'],
                            'zipcode' : d['zipcode'],
                            'lat' : d['geo']['lat'],
                            'long' : d['geo']['lat']}, 
                            ignore_index = True)
```

```python
address_df = address_df.set_index('id')
address_df
```

```python
# Continuamos con las compañías
company_dict = pd.Series.to_dict(df['company'])
company_dict
```

```python
users_ids = []
company_df = pd.DataFrame(columns = ['id', 'name', 'catchphrase', 'bs'])
company_df
```

```python
for user_id, d in company_dict.items():
    company_df = company_df.append({'id' : user_id,
                                    'name' : d['name'],
                                    'catchphrase' : d['catchPhrase'],
                                    'bs' : d['bs']},
                                    ignore_index = True)
company_df = company_df.set_index('id')
company_df
```

```python
# Ahora ya tenemos los tres dataframes, coancatenamos
dataframe_final = pd.concat([df, address_df, company_df], axis = 1)
dataframe_final
dataframe_final.drop(['address', 'company'], axis = 1)
```
